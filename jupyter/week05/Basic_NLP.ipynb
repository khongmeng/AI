```python
import spacy
from nltk.stem import PorterStemmer
```

```python
import spacy.cli
spacy.cli.download("en_core_web_md")
nlp_model = spacy.load("en_core_web_md")

#Here we're loading a collection of specialized models that do a bunch of different NLP tasks
#en: English language
#md: Medium (vs small, large, transformer)
```

```python
text = "Although the experiment appeared successful at first glance, a closer analysis revealed several subtle errors that significantly affected the final results."

doc = nlp_model(text)
```

```python
sentences = [sent.text for sent in doc.sents]
print(sentences)
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
lower = [token.lower_ for token in doc]
print(lower)
```

```python
no_stop = [token.text for token in doc if not token.is_stop and not token.is_punct]
print(no_stop)
```

```python
lemmas = [token.lemma_ for token in doc if not token.is_punct]
print(lemmas)
```

```python
stemmer = PorterStemmer()

stems = [stemmer.stem(token.text) for token in doc if token.is_alpha]
print(stems)
```

```python
pos_tags = [(token.text, token.pos_) for token in doc]
print(pos_tags)
```

```python
entities = [(ent.text, ent.label_) for ent in doc.ents]
print(entities)
```

```python
#Neat visualization of POS info + syntactic relationships
from spacy import displacy

displacy.serve(doc)
```
